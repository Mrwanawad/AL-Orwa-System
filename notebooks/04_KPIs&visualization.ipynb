```python
from  plotly.graph_objs._figure import Figure
import plotly.express as px
def apply_figure_layout( fig : Figure, title_text = '', xlabel_text = '', ylabel_text = '' ) -> Figure :
    
    xlabel_text = fig.layout.xaxis.title.text if xlabel_text == '' else xlabel_text
    ylabel_text = fig.layout.yaxis.title.text if ylabel_text == '' else ylabel_text
    title_text = fig.layout.title.text if title_text == '' else title_text  
    
    fig.update_traces(textfont_size=30)
    
    fig.update_layout(
        
        title = dict(
            pad = dict( l = 100 ), xanchor = 'left',
            text = title_text,
            font = dict( family = 'times', size = 30, color = "#0F52BA" )
        ),
        
        xaxis = dict(
            title = dict(
                text = xlabel_text, font = dict( family = 'times', size = 40, color = '#0F52BA' )
            ),
        tickfont =dict( family = 'times', size = 22, color = '#0F52BA' ),
        

        ),    
        yaxis = dict(
            title = dict(
                text = ylabel_text, font = dict( family = 'times', size = 40, color = '#0F52BA' )
            ),
        tickfont =dict( family = 'times', size = 22, color = "#0F52BA" )

        ),
        
        legend = dict(
            bordercolor = "#0F52BA",
            font = dict( family = 'times', size = 27, color = '#0F52BA', lineposition = 'under' )
        ),
        
        hoverlabel = dict(
            font = dict( family = 'times', size = 23.5, color = '#FFFFFF', lineposition = 'under' )
        ),
        
    )  
    
    fig.update_traces(
    textfont=dict(
        family="times", 
        size=28,
        )
    )
    
    return fig

    
    
category_color_map = {
    'medium' : "#5CA0AD",
    'heavy' : "#144180",
    'sitting' : "#3E772B",
    'standing' : "#AF2316"
}    
```

```python
fig =px.sunburst( df, path = [ 'النوع', 'الامبير' ], color = 'النوع', values= 'العدد',  color_discrete_sequence= ['#0F52BA'],
                 height= 600, width = 1200)
fig.update_traces( hovertemplate="<b>%{label}</b><br>العدد=%{value}<extra></extra>" )
fig = apply_figure_layout( fig )
fig
```

```python
fig =px.treemap( df, path = [ 'النوع', 'الامبير' ], color = 'النوع', values= 'العدد',  color_discrete_sequence= ['#0F52BA'],
                 height= 600, width = 1200)
fig = apply_figure_layout( fig )
fig
```
